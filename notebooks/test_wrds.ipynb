```python
import os
import wrds

# Use env vars for non-interactive runs (e.g. nbconvert)
uname = os.environ.get('WRDS_USERNAME')
pwd = os.environ.get('WRDS_PASSWORD')
if uname and pwd:
    conn = wrds.Connection(wrds_username=uname, wrds_password=pwd)
else:
    conn = wrds.Connection()

print("Connected to WRDS successfully.")
```
